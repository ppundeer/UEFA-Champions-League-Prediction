```python
### Data Cleaing : filling the missing value for the continous variables for proper data visualization

data['ShortPassing'].fillna(data['ShortPassing'].mean(), inplace = True)
data['Volleys'].fillna(data['Volleys'].mean(), inplace = True)
data['Dribbling'].fillna(data['Dribbling'].mean(), inplace = True)
data['Curve'].fillna(data['Curve'].mean(), inplace = True)
data['FKAccuracy'].fillna(data['FKAccuracy'].mean(), inplace = True)
data['LongPassing'].fillna(data['LongPassing'].mean(), inplace = True)
data['BallControl'].fillna(data['BallControl'].mean(), inplace = True)
data['HeadingAccuracy'].fillna(data['HeadingAccuracy'].mean(), inplace = True)
data['Finishing'].fillna(data['Finishing'].mean(), inplace = True)
data['Crossing'].fillna(data['Crossing'].mean(), inplace = True)

data['Weight'].fillna('200lbs', inplace = True)
data['Contract Valid Until'].fillna(2019, inplace = True)
data['Height'].fillna("5'11", inplace = True)
data['Loaned From'].fillna('None', inplace = True)
data['Joined'].fillna('Jul 1, 2018', inplace = True)
data['Jersey Number'].fillna(8, inplace = True)
data['Body Type'].fillna('Normal', inplace = True)
data['Position'].fillna('ST', inplace = True)
data['Club'].fillna('No Club', inplace = True)
data['Work Rate'].fillna('Medium/ Medium', inplace = True)
data['Skill Moves'].fillna(data['Skill Moves'].median(), inplace = True)
data['Weak Foot'].fillna(3, inplace = True)
data['Preferred Foot'].fillna('Right', inplace = True)
data['International Reputation'].fillna(1, inplace = True)
data['Wage'].fillna('€200K', inplace = True)
```

```python
data.fillna(0, inplace = True)
```

```python
def defending(data):
    return int(round((data[['Marking', 'StandingTackle', 'SlidingTackle']].mean()).mean()))

def general(data):
    return int(round((data[['HeadingAccuracy', 'Dribbling', 'Curve', 'BallControl']].mean()).mean()))

def mental(data):
    return int(round((data[['Aggression', 'Interceptions', 'Positioning', 'Vision','Composure']].mean()).mean()))

def passing(data):
    return int(round((data[['Crossing', 'ShortPassing', 'LongPassing']].mean()).mean()))

def mobility(data):
    return int(round((data[['Acceleration', 'SprintSpeed', 'Agility','Reactions']].mean()).mean()))

def power(data):
    return int(round((data[['Balance', 'Jumping', 'Stamina', 'Strength']].mean()).mean()))

def rating(data):
    return int(round((data[['Potential', 'Overall']].mean()).mean()))

def shooting(data):
    return int(round((data[['Finishing', 'Volleys', 'FKAccuracy', 'ShotPower','LongShots', 'Penalties']].mean()).mean()))
```

```python
def extract_weight(val):
    out = val.replace('lbs', '')
    return float(out)
data.Weight = data.Weight.apply(lambda x : extract_weight(x))
```

```python
def extract_value(val):
    out = val.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in out:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
data.Value = data.Value.apply(lambda x : extract_value(x))
data.Wage = data.Wage.apply(lambda x : extract_value(x))
data.Wage.head()
```

```python
features = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'LongShots', 'Dribbling',
            'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
            'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'Aggression', 'Interceptions',
            'Positioning', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 
            'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']

# Top 5 features for every position
for i,val in data.groupby('Position')[features].mean().iterrows():
    #print(i, val.nlargest(5).index)
    print('Position {} : {}, {}, {}, {}'.format(i, *tuple(val.nlargest(5).index)))
```

```python
from math import pi
plt.figure(figsize = (15,45))

p = 1
for i,val in data.groupby('Position')[features].mean().iterrows():
    top_features = dict(val.nlargest(5))
    ctg = top_features.keys()
    N = len(ctg)
    
    # We need to repeat the first value to close the circular graph:
    values = list(top_features.values())
    values += values[:1]
    
    angle = [n / float(N) * 2 * pi for n in range(N)]
    angle += angle[:1]
    
    ax = plt.subplot(9, 3, p, polar = True)
    
    plt.xticks(angle, ctg, color = 'grey', size = 8)
    
    ax.set_rlabel_position(0)
    plt.yticks([25,50,75], ['25','50','75'], color = 'grey')
    plt.ylim(0,100)
    
    ax.plot(angle, values, linewidth=1, linestyle='solid')
    ax.fill(angle, values, 'b', alpha=0.1)
    plt.title(i, size = 12)
    
    p += 1
```

### Player Performance graph

```python
features = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'LongShots', 'Dribbling',
            'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
            'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'Aggression', 'Interceptions',
            'Positioning', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 
            'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
vect = []
pos = []
for i,val in data.groupby('Position')[features].mean().iterrows():
    xl = val.nlargest(5).index
    vect.append(xl)
    pos.append(i)

feat = pd.DataFrame(vect,pos).T

from math import pi
plt.figure(figsize = (7,7))

def playerdataplot(x):
    temp = data_selected.loc[x,:]
    topfeature = feat[temp.Position]
    
    values = []
    for i in topfeature:
        values.append(float(temp[i]))
    values += values[:1]
    #values = list(top_features.values())

    angle = [n / float(5) * 2 * pi for n in range(5)]
    angle += angle[:1]
    
    ax = plt.subplot(111, polar = True)
    
    plt.xticks(angle, topfeature, color = 'grey', size = 12)
    ax.set_rlabel_position(0)
    plt.yticks([25,50,75], ['25','50','75'], color = 'grey')
    plt.ylim(0,100)
    
    ax.plot(angle, values, linewidth=1, linestyle='solid')
    ax.fill(angle, values, 'b', alpha=0.1)
    plt.title(temp.Name + ' - ' + temp.Club + ' : ' + temp.Position, size = 15)
    
playerdataplot(32)
```
